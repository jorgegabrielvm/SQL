```python
import pandas as pd

df = pd.read_csv('MOVIES_DATES.csv')

df
```

```python

df['Year'] = pd.to_datetime(df['Year'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')

df
```

```python
df['Year'] = df['Year'].str.split('-').str[0]

df
```

```python
df.to_csv('new_file_MOVIES_DATE.csv', index=False)
```
